```
import requests

def download_file(url, local_filename):
    with requests.get(url, stream=True) as response:
        # Controlla se la richiesta HTTP ha avuto successo (status code 200)
        if response.status_code == 200:
            # Apri il file in modalità binaria per la scrittura
            with open(local_filename, 'wb') as file:
                # Scrivi i dati del file nel file locale
                for chunk in response.iter_content(chunk_size=128):
                    file.write(chunk)
            file.close()
            print(f"Download completato. File salvato come {local_filename}")
        else:
            print(f"Errore durante il download. Status code: {response.status_code}")

# download test file
url = 'https://drive.google.com/file/d/1UGfd-iE-qLtFWO6B7WWyI-fHvbC2_rSk/view?usp=drive_link'
local_filename = 'test.wav'
download_file(url, local_filename)
```

```
from transformers import pipeline

# Initializing a speech recognition pipeline
speech_recognizer = pipeline("automatic-speech-recognition")
```

```
!pip install -q datasets
```

```
from datasets import load_dataset

# Loading a dataset named "PolyAI/minds14" with the English (en-US) split
dataset = load_dataset("PolyAI/minds14", name="en-US", split="train")
```

```
dataset.features
```

```
from datasets import Audio

# Casting the "audio" column to the Audio data type
dataset = dataset.cast_column("audio", Audio(sampling_rate=speech_recognizer.feature_extractor.sampling_rate))
```

```
result = speech_recognizer(dataset[:2]["audio"])
print([d["text"] for d in result])
```
